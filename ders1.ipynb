```python
import numpy as np        # Sayısal işlemler
import pandas as pd       # Veri analizi
import matplotlib.pyplot as plt  # Görselleştirme
```

```python
data = {
    "isim": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "yas": [25, 30, None, 40, 29],
    "maas": [5000, 7000, 6500, None, 7200]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Eksik veri tespiti
df.isnull()
df.isnull().sum()
```

```python
#Eksik veriyi doldurma
df["yas"] = df["yas"].fillna(df["yas"].mean())
df["maas"] = df["maas"].fillna(df["maas"].median())

df
```

```python
plt.hist(df["maas"])
plt.title("Maas Dağılımı")
plt.xlabel("Maas")
plt.ylabel("Frekans")
plt.show()
```

```python
plt.scatter(df["yas"], df["maas"])
plt.title("Yaş vs Maaş")        
plt.xlabel("Yaş")
plt.ylabel("Maaş")
plt.show()
```
